```python
from data_list import ImageList
import torch
import pre_process as prep
import torchvision
import matplotlib.pyplot as plt
import numpy as np
```

```python
def imshow(img):
#     img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.figure(figsize=(16,16))
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.axis('off')
    plt.show()
```

### DomainNet

```python
dataset = "clipart"

loadfile = "data/visDA_full/{}_test.txt".format(dataset)

dataset_list = ImageList("/newfoundland/tarun/datasets/Adaptation/visDA/", open(loadfile).readlines(),
                                transform=prep.image_test(resize_size=256, crop_size=224))

dataset_loaders = torch.utils.data.DataLoader(dataset_list, batch_size=256, shuffle=True, num_workers=0)

images, _ = next(iter(dataset_loaders))

imshow(torchvision.utils.make_grid(images, nrow=16))
```
