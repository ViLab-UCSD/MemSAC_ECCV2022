```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
```

```python
def getaccs(df):
    df["l1names"] = df["l1"].map(fineMapping)
    df["l2names"] = df["l2"].map(coarse_mapping)

    #define conditions
    conditions = [df['l1'] == df['pred'], 
                  df['l1'] != df['pred']]

    #define choices
    choices = [1,0]


    df["match"] = np.select(conditions , choices )
    
    return df
```

```python
fileName = "/home/tarun/ILADA/snapshot/from_nautilus/CDAN_domainNet_painting_clipart_QS16384_sc0/acct.txt"

df = pd.read_csv(fileName, delimiter="\t", header=0, names=["iter" , "l1" , "l2" , "pred"])
df_cdan = getaccs(df)
accs_cdan = 100*(df_cdan.groupby("l2names").mean()["match"].values)

fileName = fileName.replace("sc0" , "sc0-1")
df = pd.read_csv(fileName, delimiter="\t", header=0, names=["iter" , "l1" , "l2" , "pred"])
df_ila = getaccs(df)
accs_ila = 100*(df_ila.groupby("l2names").mean()["match"].values)

import matplotlib.pyplot as plt
index = category_counts

counts = list(category_counts.values())
idx = sorted(range(len(counts)), key=lambda k:counts[k], reverse=True)

accs_ila, accs_cdan = accs_ila[idx], accs_cdan[idx]
values_diff = accs_ila - accs_cdan
colors = ["green" if v>0 else "red" for v in values_diff]

fig, ax = plt.subplots(figsize=(14,6))
plt.scatter(range(len(index)) , accs_cdan, s=75, facecolors='none', edgecolors='k', label="cdan")
# plt.bar(range(len(index)) , np.array(list(category_counts.values()))[idx])
for i in range(len(colors)):
    if colors[i] == "green":
        plt.arrow(i , accs_cdan[i] , 0 , values_diff[i], width=0.06, head_width=0.5, color="green" )
    else:
        plt.arrow(i , accs_cdan[i] , 0 , values_diff[i], width=0.06, head_width=0.5, color="red" )
#     plt.quiver(range(len(index)), accs_cdan, [0]*len(index), values_diff, color=colors, width=0.004)

plt.yticks([30,40,50,60,70,80,90], fontsize=16)
ks = list(category_counts.keys())
# labels = [ks[i]+ " ("+str(category_counts[ks[i]])+")" for i in idx]
labels = [ks[i] for i in idx]
plt.xticks(range(len(index)) , labels, rotation=90, fontsize=20)

# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_linewidth(4.5)
ax.spines['bottom'].set_linewidth(4.5)
# plt.legend()

# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

ax.yaxis.grid(True)
plt.grid(which="major", color="black", linestyle='--', linewidth=0.15)

# plt.xlabel("Coarse Categories in domainnet", fontsize=24)
plt.ylabel("Accuracy (%)", fontsize=20)
# plt.title("Accuracy gain vs. category population.", fontsize=20)

# plt.legend(fontsize=20, shadow=True, loc="upper center", ncol=3, handlelength=3)

# plt.savefig("ClassWise_Acc_SP.pdf", format='pdf', dpi=1200, bbox_inches='tight')

fig.show();
```
