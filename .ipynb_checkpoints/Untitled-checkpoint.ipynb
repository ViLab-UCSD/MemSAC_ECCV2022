```python
import numpy as np
from scipy.special import comb, factorial
```

```python
def fn(x,N):
    return (comb(N,x))/((N**x)*factorial(x))
```

```python
N = 5
vals_5 = np.array([fn(i,N) for i in np.arange(1,N+1)])

N = 10
vals_10 = np.array([fn(i,N) for i in np.arange(1,N+1)])

N = 15
vals_20 = np.array([fn(i,N) for i in np.arange(1,N+1)])
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.plot(1./vals_5 , label="N5")
plt.plot(1./vals_10 , label="N10")
plt.plot(1./vals_20 , label="N15")
plt.yscale("log")
plt.legend()
fig.show()
```

```python
import numpy as np
def invEuc(x):
    return 1./(1+x)

def GaussEuc(x, tau):
    return np.exp(-x)
```

```python
distVals = np.arange(5,12,0.005)
invEuc_sim = invEuc(distVals)
GaussEuc_sim_10 = GaussEuc(distVals, tau=10)
# GaussEuc_sim_007 = GaussEuc(distVals, tau=0.07)
GaussEuc_sim_1 = GaussEuc(distVals, tau=1)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.plot(distVals, invEuc_sim**2, label="invEuc")
plt.plot(distVals, GaussEuc_sim_1 , label="Exp. Gaussian")
# plt.plot(distVals, GaussEuc_sim_007 , label="Gauss_0.07")
# plt.plot(distVals, GaussEuc_sim_10 , label="Gauss_10")
# plt.xticks(distVals)
plt.xlabel("Distance between features")
plt.ylabel("Gradient of Similarity value.")
plt.legend()
fig.show()
```
