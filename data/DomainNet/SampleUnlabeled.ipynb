```python
import os
import pandas as pd
import numpy as np
```

```python
target_file = "clipart_train_50.txt"
```

```python
class_file_list = {str(i):[] for i in range(345)}
with open(target_file, "r") as fh:
    for line in fh.readlines():
        class_id = line.strip().split(" ")[-1]
        class_file_list[class_id].append(line)
```

```python
percentages = [5,10,25,50,75,90]
```

```python
for p in percentages:
    new_file_list = []
    for v in class_file_list.values():
        sampled_len = max(1 , int(len(v)*p/100))
        new_file_list.extend(v[:sampled_len])
    with open("clipart_train_%d.txt"%p , "w") as fh:
        for f in new_file_list:
            fh.write(f)
```
